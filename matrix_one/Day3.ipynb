```
df_usd = df[ df.prices_currency == 'USD'].copy()
```

```
df_usd.shape
```

```
df_usd['prices_amountmin'] = df_usd.prices_amountmin.astype(np.float)
df_usd['prices_amountmin'].hist()
```

```
filter_max = np.percentile( df_usd['prices_amountmin'], 99 )
filter_max
```
